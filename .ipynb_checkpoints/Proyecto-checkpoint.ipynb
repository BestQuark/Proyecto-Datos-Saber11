```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sp
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
datos = pd.read_csv('datos_SPP.csv') 
df = datos.filter(['puntaje_saber11', 'mujer', 'edad', 'minoria', 'beneficiario_spp', 'miembros_familia','col_privado','icfes_educ_madre', 'estrato'])
df = df.dropna()
educ_madre = {'Primaria': 1, 'Secundaria': 2, 'Tecnica': 3, 'Universitaria':4}
df = df.replace(educ_madre)
```

```python
df['mujer*edad'] = df['mujer']*df['edad']
df['mujer*minoria'] = df['mujer']*df['minoria']
df['mujer*beneficiario_spp'] =  df['mujer']*df['beneficiario_spp']
df['mujer*miembros_familia'] = df['mujer']*df['miembros_familia']
df['mujer*col_privado'] = df['mujer']*df['col_privado']
df['mujer*icfes_educ_madre'] = df['mujer']*df['icfes_educ_madre']
df['mujer*estrato'] = df['mujer']*df['estrato']

# df['estrato*edad'] = df['estrato']*df['edad']
# df['estrato*minoria'] = df['estrato']*df['minoria']
# df['estrato*spp'] = df['estrato']*df['beneficiario_spp']
# df['estrato*miembrosFam'] = df['estrato']*df['miembros_familia']
# df['estrato*colpriv'] = df['estrato']*df['col_privado']
# df['estrato*madre'] = df['estrato']*df['icfes_educ_madre']

# df['col_privado*edad'] = df['col_privado']*df['edad']
# df['col_privado*minoria'] = df['col_privado']*df['minoria']
# df['col_privado*spp'] = df['col_privado']*df['beneficiario_spp']
# df['col_privado*miembrosFam'] = df['col_privado']*df['miembros_familia']
# df['col_privado*madre'] = df['col_privado']*df['icfes_educ_madre']

# df['madre*edad'] = df['icfes_educ_madre']*df['edad']
# df['madre*minoria'] = df['icfes_educ_madre']*df['minoria']
# df['madre*spp'] = df['icfes_educ_madre']*df['beneficiario_spp']
# df['madre*miembrosFam'] = df['icfes_educ_madre']*df['miembros_familia']
```

```python
columnasInic = df.columns
for indi, i in enumerate(columnasInic[1:]):
    for indj, j in enumerate(columnasInic[1:]):
        if i!=j and indi<indj:
            df[str(i+'*'+j)] = df[i]*df[j]
```

```python
mujeresPriv = df.loc[(df['mujer'] == 1) & (df['col_privado']==1)]
hombresPriv = df.loc[(df['mujer'] == 0) & (df['col_privado']==1)]
mujeresPubli = df.loc[(df['mujer'] == 1) & (df['col_privado']==0)]
hombresPubli = df.loc[(df['mujer'] == 0) & (df['col_privado']==0)]
mujeres = df.loc[df['mujer'] == 1]
hombres = df.loc[df['mujer'] == 0]
```

```python
print("mujeresPriv \n promedio: ", mujeresPriv["puntaje_saber11"].mean(), "\n std :", mujeresPriv["puntaje_saber11"].std())
print("hombresPriv \n promedio: ", hombresPriv["puntaje_saber11"].mean(),"\n std :", hombresPriv["puntaje_saber11"].std())
print("mujeresPubli \n promedio: ", mujeresPubli["puntaje_saber11"].mean(), "\n std :", mujeresPubli["puntaje_saber11"].std())
print("hombresPubli \n promedio: ", hombresPubli["puntaje_saber11"].mean(),"\n std :", hombresPubli["puntaje_saber11"].std())
```

```python
sns.set()
f, ax = plt.subplots(1,2, figsize = (15,5))
ax[0].hist(mujeresPriv["puntaje_saber11"], bins=50, alpha=0.5, label = 'F C.Privado')
ax[0].hist(hombresPriv["puntaje_saber11"], bins=50, alpha=0.5, label = 'M C.Privado')
ax[1].hist(mujeresPubli["puntaje_saber11"], bins=50, alpha=0.5, label = 'F C.Público')
ax[1].hist(hombresPubli["puntaje_saber11"], bins=50, alpha=0.5, label = 'M C.Público')
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
muMujer, stdMujer = sp.norm.fit(mujeres["puntaje_saber11"])
muHombre, stdHombre = sp.norm.fit(hombres["puntaje_saber11"])

sns.set()
plt.hist(mujeres["puntaje_saber11"], bins=100, density=True, alpha=0.5, label = 'F')
plt.hist(hombres["puntaje_saber11"], bins=100, density=True, alpha=0.5, label = 'M')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pMujer = sp.norm.pdf(x, muMujer, stdMujer)
pHombre = sp.norm.pdf(x, muHombre, stdHombre)
plt.plot(x, pMujer, 'k', linewidth=2, c= 'blue')
plt.plot(x, pHombre, 'k', linewidth=2, c = 'orange')
plt.legend()
plt.show()
```

```python
fit_alphaM, fit_locM, fit_betaM= sp.gamma.fit(mujeres["puntaje_saber11"])
fit_alphaH, fit_locH, fit_betaH= sp.gamma.fit(hombres["puntaje_saber11"])

sns.set()
plt.hist(mujeres["puntaje_saber11"], bins=50, density=True, alpha=0.5, label = 'F')
plt.hist(hombres["puntaje_saber11"], bins=50, density=True, alpha=0.5, label = 'M')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pMujer = sp.gamma.pdf(x, fit_alphaM, loc=fit_locM , scale=fit_betaM)
pHombre = sp.gamma.pdf(x, fit_alphaH, loc=fit_locH , scale=fit_betaH)
plt.plot(x, pMujer, 'k', linewidth=2, c= 'blue')
plt.plot(x, pHombre, 'k', linewidth=2, c = 'orange')
plt.legend()
plt.show()
```

```python
muMujerP, stdMujerP = sp.norm.fit(mujeresPriv["puntaje_saber11"])
muHombreP, stdHombreP = sp.norm.fit(hombresPriv["puntaje_saber11"])
plt.hist(mujeresPriv["puntaje_saber11"], bins=50, density=True, alpha=0.5, label = 'F')
plt.hist(hombresPriv["puntaje_saber11"], bins=50, density=True, alpha=0.5, label = 'M')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pMujerP = sp.norm.pdf(x, muMujerP, stdMujerP)
pHombreP = sp.norm.pdf(x, muHombreP, stdHombreP)
plt.plot(x, pMujerP, 'k', linewidth=2, c= 'blue')
plt.plot(x, pHombreP, 'k', linewidth=2, c = 'orange')
plt.legend()
plt.show()
```
